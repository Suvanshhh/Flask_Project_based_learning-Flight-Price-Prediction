## 4. Data Preprocessing

```python
dt_cols = ["date_of_journey","dep_time", "arrival_time"]

num_cols = ["duration", "total_stops"]

cat_cols = [col for col in X_train.columns if (col not in dt_cols) and (col not in num_cols)]

```

```python
cat_cols
```

```python
num_transformer = Pipeline(steps = [("imputer", SimpleImputer(strategy = "median")),
                            ("scaler", StandardScaler())
                            ])
cat_transformer = Pipeline(steps = [("imputer", SimpleImputer(strategy = "most_frequent")),
                            ("encoder", OneHotEncoder(sparse_output = False))
                            ])
doj_transformer = Pipeline(steps = [("imputer", SimpleImputer(strategy = "most_frequent")),
                                    ("extracter", DatetimeFeatures(features_to_extract = ["month", "week", "day_of_week", "day_of_month"], format = "mixed")),
                                    ("scaler", StandardScaler())
                                   
                                   ])

time_transformer = Pipeline(steps = [("imputer", SimpleImputer(strategy = "most_frequent")),
                                    ("extracter", DatetimeFeatures(features_to_extract = ["hour", "minute"], format = "mixed")),
                                    ("scaler", StandardScaler())
                                    ])
```

```python
preprocessor = ColumnTransformer(transformers = [("num", num_transformer, num_cols),
                                                ("cat", cat_transformer, cat_cols),
                                                 ("doj", doj_transformer, ["date_of_journey"]),
                                                 ("time", time_transformer, ["dep_time", "arrival_time"])
                                                ])
```

```python
preprocessor.fit_transform(X_train)
```

```python
def plot_curves(sizes, train_scores, std_scores, label, ax):
    ax.plot(
        sizes,
        mean_scores,
        marker = "o",
        label = label
    )
    ax.fill_between(
     x=sizes,
    y1 = mean_scores - std_scores,
    y2 = mean_scores - std_scores,
    alpha=0.5
)
    

```

```python
def plot_learning_curves(name, algorithm, X_data, y_data, preprocessor, figsize=(12, 4)):
    from sklearn.pipeline import Pipeline
    from sklearn.model_selection import learning_curve
    import numpy as np
    import matplotlib.pyplot as plt

    # Define the model pipeline
    model = Pipeline(steps=[
        ("pre", preprocessor),
        ("alg", algorithm),
    ])

    # Compute learning curves
    train_sizes, train_scores, test_scores = learning_curve(
        estimator=model,
        X=X_data,
        y=y_data,
        cv=3,
        scoring="r2",
        n_jobs=-1,
        random_state=42
    )

    # Calculate mean and standard deviation for training scores
    mean_train_scores = np.mean(train_scores, axis=1)
    std_train_scores = np.std(train_scores, axis=1)
    train_score = f"{mean_train_scores[-1]:.2f} +/- {std_train_scores[-1]:.2f}"

    # Calculate mean and standard deviation for test scores
    mean_test_scores = np.mean(test_scores, axis=1)
    std_test_scores = np.std(test_scores, axis=1)
    test_score = f"{mean_test_scores[-1]:.2f} +/- {std_test_scores[-1]:.2f}"

    # Plot learning curves
    fig, ax = plt.subplots(figsize=figsize)

    ax.plot(train_sizes, mean_train_scores, 'o-', color="r", label=f"Train ({train_score})")
    ax.fill_between(train_sizes, mean_train_scores - std_train_scores, mean_train_scores + std_train_scores, alpha=0.1, color="r")

    ax.plot(train_sizes, mean_test_scores, 'o-', color="g", label=f"Test ({test_score})")
    ax.fill_between(train_sizes, mean_test_scores - std_test_scores, mean_test_scores + std_test_scores, alpha=0.1, color="g")

    ax.set(xlabel="Training Set Size", ylabel="R-Square", title=name)
    ax.legend(loc="lower right")

    plt.show()

# Example usage
# plot_learning_curves("Model Name", your_algorithm, X_data, y_data, preprocessor)
```

```python
# Example data and preprocessor (make sure to replace with your actual data and preprocessor)
from sklearn.datasets import make_regression
from sklearn.preprocessing import StandardScaler

# Generate some example data
X_data, y_data = make_regression(n_samples=100, n_features=20, noise=0.1, random_state=42)

# Define a simple preprocessor
preprocessor = StandardScaler()

# Example algorithms (make sure to replace with your actual algorithms)
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

algorithms = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Support Vector Machine": SVR(),
    "Random Forest": RandomForestRegressor(n_estimators=10)
    }

# Plot learning curves for each algorithm
for name, alg in algorithms.items():
    plot_learning_curves(name, alg, X_data, y_data, preprocessor)
```
